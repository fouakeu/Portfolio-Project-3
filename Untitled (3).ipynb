# Data Preprocessing

Split DataFrame in train and test

```python
# input variable
X = cancer_df.drop(['target'], axis = 1)
X.head(6)
```

```python
# output variable
y = cancer_df['target']
y.head(6)
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
# Feature scaling
# Keep only numeric columns
X_train_numeric = X_train.select_dtypes(include=['float64', 'int64'])
X_test_numeric = X_test[X_train_numeric.columns]  # make sure columns match

# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train_numeric)
X_test_sc = sc.transform(X_test_numeric)
```

# Support Vector Classifier

```python
# Support vector classifier
# Keep only numeric columns
X_train_numeric = X_train.select_dtypes(include=['float64', 'int64'])
X_test_numeric = X_test[X_train_numeric.columns]  # ensure columns match

# Feature scaling (optional but recommended for SVC)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train_numeric)
X_test_sc = sc.transform(X_test_numeric)

# Train SVC
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train_sc, y_train)

# Predict
y_pred_svc = svc_classifier.predict(X_test_sc)

# Evaluate accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_svc)
print("SVC Accuracy:", accuracy)
```

# Logistic Regression

```python
# Select only numeric columns
X_train_numeric = X_train.select_dtypes(include=['float64', 'int64'])
X_test_numeric = X_test[X_train_numeric.columns]  # make sure columns match

# Feature scaling (optional but recommended)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train_numeric)
X_test_sc = sc.transform(X_test_numeric)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_classifier = LogisticRegression(random_state=51, penalty='l1', solver='liblinear')
lr_classifier.fit(X_train_sc, y_train)

y_pred_lr = lr_classifier.predict(X_test_sc)
accuracy = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#Sélection des colonnes numériques
# Supprimer toutes les colonnes de type 'Interval' ou non numériques
numeric_cols = cancer_df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Exclure la colonne target de X
X = cancer_df[numeric_cols].drop('target', axis=1)
y = cancer_df['target']

#Train/Test split (avec random_state pour reproductibilité)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=99, stratify=y
)

#Standardisation
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

#Logistic Regression avec L1
lr_classifier = LogisticRegression(random_state=51, penalty='l1', solver='liblinear')
lr_classifier.fit(X_train_sc, y_train)

#Prédiction
y_pred = lr_classifier.predict(X_test_sc)

#Évaluation
accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression Accuracy (scaled, L1):", accuracy)
```

# Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

adb_classifier = AdaBoostClassifier(
    DecisionTreeClassifier(criterion='entropy', random_state=200),
    n_estimators=2000,
    learning_rate=0.1,
    algorithm='SAMME',   
    random_state=1
)

adb_classifier.fit(X_train, y_train)
y_pred_adb = adb_classifier.predict(X_test)
print("AdaBoost Accuracy:", accuracy_score(y_test, y_pred_adb))
```

# XGBoost Parameter Tuning Randomized Search

```python
# XGBoost classifier most required parameters
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ] 
}
```

```python
# Randomized Search
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(xgb_classifier, param_distributions=params, scoring= 'roc_auc', n_jobs= -1, verbose= 3)
random_search.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Training XGBoost classifier with best parameters
xgb_classifier_pt = XGBClassifier(
    base_score=0.5, booster='gbtree', colsample_bytree=0.4,
    colsample_bylevel=1, colsample_bynode=1, gamma=0.2,
    learning_rate=0.1, max_delta_step=0, max_depth=15,
    min_child_weight=1, n_estimators=100, n_jobs=1,
    objective='binary:logistic', random_state=0,
    reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
    subsample=1, verbosity=1,  # keep verbosity=1 for info
    missing=np.nan  # <-- use np.nan instead of None
)

# Fit the model
xgb_classifier_pt.fit(X_train, y_train)

# Predict on the test set
y_pred_xgb_pt = xgb_classifier_pt.predict(X_test)

# Evaluate accuracy
print("XGBoost Accuracy (best parameters):", accuracy_score(y_test, y_pred_xgb_pt))
```
